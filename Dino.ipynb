```python
import os
import shutil
import json
from sklearn.model_selection import train_test_split

# Paths
dataset = '/content/DINO/datasets/Pedestrian_dataset_for_internship_assignment'
annotations_path = '/content/random_sample_mavi_2_gt.json'  # Path to your single annotation file
output_dir = '/content/DINO/datasets/Pedestrian_dataset_for_internship_assignment'
train_dir = os.path.join(output_dir, 'train2017')
val_dir = os.path.join(output_dir, 'val2017')
annotations_dir = os.path.join(output_dir, 'annotations')

# Ensure directories exist
os.makedirs(train_dir, exist_ok=True)
os.makedirs(val_dir, exist_ok=True)
os.makedirs(annotations_dir, exist_ok=True)

# Load images and partition into train/val
image_files = [f for f in os.listdir(dataset) if f.endswith(('.png', '.jpg', '.jpeg'))]
train_images, val_images = train_test_split(image_files, test_size=41, train_size=160, random_state=42)

# Move images to respective directories
for img in train_images:
    shutil.move(os.path.join(dataset, img), os.path.join(train_dir, img))
for img in val_images:
    shutil.move(os.path.join(dataset, img), os.path.join(val_dir, img))

# Load original annotation file
with open(annotations_path, 'r') as f:
    data = json.load(f)

# Get image details for train and val splits
train_image_names = {img.split('.')[0] for img in train_images}
val_image_names = {img.split('.')[0] for img in val_images}

# Filter images and annotations
train_images_data = [img for img in data['images'] if img['file_name'].split('.')[0] in train_image_names]
val_images_data = [img for img in data['images'] if img['file_name'].split('.')[0] in val_image_names]

train_image_ids = {img['id'] for img in train_images_data}
val_image_ids = {img['id'] for img in val_images_data}

train_annotations = [ann for ann in data['annotations'] if ann['image_id'] in train_image_ids]
val_annotations = [ann for ann in data['annotations'] if ann['image_id'] in val_image_ids]

# Add categories (ensure the 'categories' section has 'pedestrian' category as per DINO requirements)
categories = [{"id": 1, "name": "pedestrian", "supercategory": "pedestrian"}]

# Create COCO-style data structures
train_data = {"images": train_images_data, "annotations": train_annotations, "categories": categories}
val_data = {"images": val_images_data, "annotations": val_annotations, "categories": categories}

# Save train and validation JSON files
with open(os.path.join(annotations_dir, 'instances_train2017.json'), 'w') as f:
    json.dump(train_data, f)
with open(os.path.join(annotations_dir, 'instances_val2017.json'), 'w') as f:
    json.dump(val_data, f)

print("Dataset partitioned and COCO-style annotation files created.")
```

```python
import json
import matplotlib.pyplot as plt

# Initialize lists to store the metrics over epochs
train_loss = []
train_class_error = []
train_lr = []

# Open and read each line in log.txt
with open('/content/logs/DINO/R50-MS4/log.txt', 'r') as f:
    for line in f:
        # Load the line as a JSON object
        data = json.loads(line.strip())

        # Append the metrics for each epoch to their respective lists
        train_loss.append(data.get("train_loss", None))
        train_class_error.append(data.get("train_class_error", None))
        train_lr.append(data.get("train_lr", None))

# Step 3: Plot the metrics
# Plot the training loss over epochs
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(train_loss, label='Training Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training Loss Curve')
plt.legend()

# # Plot the training class error over epochs
# plt.subplot(1, 2, 2)
# plt.plot(train_class_error, label='Training Class Error', color='orange')
# plt.xlabel('Epoch')
# plt.ylabel('Class Error (%)')
# plt.title('Training Class Error Curve')
# plt.legend()

# Display the plots
plt.tight_layout()
plt.show()
```

```python
import json
import matplotlib.pyplot as plt

# Initialize lists for each type of loss
train_loss_ce = []
train_loss_bbox = []
train_loss_giou = []

# Intermediate losses (if available)
train_loss_ce_0 = []
train_loss_bbox_0 = []
train_loss_giou_0 = []

# Open and read each line in log.txt
with open('/content/logs/DINO/R50-MS4/log.txt', 'r') as f:
    for line in f:
        data = json.loads(line.strip())

        # Append main losses
        train_loss_ce.append(data.get("train_loss_ce", None))
        train_loss_bbox.append(data.get("train_loss_bbox", None))
        train_loss_giou.append(data.get("train_loss_giou", None))

        # Append intermediate losses
        train_loss_ce_0.append(data.get("train_loss_ce_0", None))
        train_loss_bbox_0.append(data.get("train_loss_bbox_0", None))
        train_loss_giou_0.append(data.get("train_loss_giou_0", None))

# Plot main losses
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(train_loss_ce, label='Classification Loss (CE)')
plt.plot(train_loss_bbox, label='Bounding Box Loss')
plt.plot(train_loss_giou, label='GIoU Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Main Losses')
plt.legend()

# Plot intermediate losses
plt.subplot(1, 2, 2)
plt.plot(train_loss_ce_0, label='Classification Loss (CE) - Intermediate')
plt.plot(train_loss_bbox_0, label='Bounding Box Loss - Intermediate')
plt.plot(train_loss_giou_0, label='GIoU Loss - Intermediate')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Intermediate Losses')
plt.legend()

plt.tight_layout()
plt.show()
```
